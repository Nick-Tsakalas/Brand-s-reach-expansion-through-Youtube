```python
import pandas as pd
```

```python
with open('Global YouTube Statistics.csv', encoding="utf8", errors='ignore') as f:
    df = pd.read_csv(f)
df
```

```python
df['Country'].unique()
```

```python
eulist = ['Russia', 'Andorra',
       'United Kingdom', 'Spain', 'Turkey', 'Netherlands', 'Italy',
       'Germany', 'France', 'Sweden', 'Ukraine', 'Latvia',
       'Switzerland', 'Finland']
```

```python
northamericalist = [ 'United States', 'Canada']
```

```python
asialist = ['India', 'Japan', 'South Korea',
      'Pakistan', 'Philippines', 'Thailand', 'Indonesia',
       'Jordan', 'Singapore', 'Australia', 'Afghanistan', 'Vietnam', 'Malaysia', 'China', 'Iraq', 'Kuwait',
        'Bangladesh','Saudi Arabia','United Arab Emirates' ]
```

```python
southamericalist = ['Brazil', 'Argentina', 'Chile', 'Cuba',
       'El Salvador', 'Colombia', 'Barbados', 'Mexico', 'Venezuela', 'Ecuador', 'Peru', 'Samoa']
```

```python
africalist = [ 'Morocco', 'Egypt']
```

```python
len(eulist)
```

```python
len(northamericalist)
```

```python
len(asialist)
```

```python
len(southamericalist)
```

```python
def regionx(i):
    if i in eulist:
        return 'eu'
    elif i in northamericalist:
        return 'na'
    elif i in southamericalist:
        return 'sa'
    elif i in africalist:
        return 'af'
    elif i in asialist:
        return 'as'
    else:
        return None
```

```python
df['Country'].apply(lambda x: regionx(x))
```

```python
df['region'] = df['Country'].apply(lambda x: regionx(x))
```

```python
df.head(10)
```

```python
df.to_csv('df2.csv')
```

```python
with open('df2.csv', encoding="utf8", errors='ignore') as f:
    df_new = pd.read_csv(f)
df_new
```

```python
dff = df_new.copy()
```

```python
dff['Urban_percentage'] = dff['Urban_population']/dff['Population']*100
```

```python
dff.groupby('region')['Urban_percentage'].mean().reset_index().plot.bar(x='region')
```

```python
def urbanity(x):
    if x<=40:
        return 'not urbanised'
    elif x>40 and x<70 :
        return 'somewhat urbanised'
    else:
        return 'fully urbanised'
```

```python
dff['how_much_urban']  = dff['Urban_percentage'].apply(lambda x:urbanity(x) )
```

```python
dff.head(10)
```

```python
dff.groupby('how_much_urban')['how_much_urban'].count()
```

```python
dff['Views per Subs'] = dff['video views']/dff['subscribers']
```
